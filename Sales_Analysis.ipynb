# Now Doing Exploratory Data Analysis

## Gender

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

* Females are the majority of buyers as can be seen from the graph 

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

* Women also have higher purchasing power than men.

## Age

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

* From both above graphs we can infer that the majority of sales is among the age group of 26-35 yrs Female.

```python
# def distplot(data, x, ax):
#     bins = range(-10, 110, 10)
#     sns.displot(data[x], bins=bins, ax=ax)
    
# fig, ax = plt.subplots(1, 3, figsize=(27, 7))
# for idx, y in enumerate(['Age', 'Orders']):
#     distplot(df, y, ax[idx])
sns.displot(data=df,x='Amount')
```

```python
sns.jointplot(x='Age',y='Amount',data=df,kind='kde',hue='Gender')
```

```python
sns.jointplot(x='Age',y='Amount',data=df,kind='scatter')
```

* There is no linear relationship between Amount and Age. Majority of Buyers are 28-30 years old Females. The majority of amounts are less than 10000 (peak at around 6000-7000 in terms of sale/count). There are also outliers for amount greater than 15000 and age greater than 60. 

## State

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sns.set(rc={'figure.figsize':(22,5)})
sns.boxplot(data=df,x='State',y='Amount',gap=0.5)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

* From the above graphs it can be seen that the majority of orders as well as the total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka in decreasing manner. Maharahstra has highest median Amount Value while Uttarakhand and Jharkhand have lowest.

## Marital Status

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

* Most of the purchases are from Women(unmarried) and they have high purchasing power.

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Occupation',y='Amount')
```

```python
sns.violinplot(data=df,x='Occupation',y='Amount')

```

* Most of the Buyers are from IT, Healthcare and Aviation sector and they also have high purchasing power among other occupations.

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount')
```

* We can see that while clothing has the highest sales in terms of amount Food is higher so on considering both of the above graphs we can infer that the most sold products and the most amount giving products are Food, Clothing and Electronics in the respective order.

```python
sales_state=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Product_ID',y='Orders')
```

* The most Orders of are product P00265242

### Zone

```python
ax=sns.boxplot(data=df,x='Zone',y='State')
# df.head(10)
```

```python
sns.boxplot(data=df,x='Zone',y='Amount')
```

* Central Zone has the widest distribution, indicating a high variability in the data for states within this zone. States range from Uttar Pradesh at the Upper extreme to Madhya Pradesh at the Lower extreme. 

```python
sns.violinplot(data=df,x='Zone',y='State')
plt.title('State Distribution across Zones')
```

* The median of Central Zone is Delhi, Northern is Haryana, Southern is Karnataka, Eastern is Bihar, Western is Maharashtra. Also majority of the states lies in Central Zone and Northern Zone. 

```python
sns.violinplot(data=df,x='Zone',y='Amount')
plt.title('Amount Distribution across Zones')
```

* The median of all zones is almost the same (lies in 5000-10000 Amount range) suggesting that amount is same across all zones. The Amount range is also amost same in all zones.

```python
corr=df[['Age','Marital_Status','Orders','Amount']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',center=0)
```

* There is no correlation (all are very weak correaltion) suggesting no linear relationship.
